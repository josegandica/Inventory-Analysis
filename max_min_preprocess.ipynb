```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('main.csv') 
```

```python
#df = df.rename(columns={"to_action":"action"})
##df['to_location_count'] = pd.NA
##df['from_location_count'] = pd.NA
df['quantity'] = df['quantity'].str.replace(',','')
df['quantity'] = pd.to_numeric(df['quantity'])
df
```

```python
n = df.nunique(axis=0)
n
```

```python
def stock(row):
    if row['action'] == 'ADD':
        return row['quantity'] * 1
    elif row['action'] == 'DELETE':
        return row['quantity'] * -1
    elif row['action'] == 'TRANSFER':
        return row['quantity'] * -1

df['moved_quantity_from'] = df.apply(stock, axis=1)

df
```

```python
def stock(row):
    if row['action'] == 'ADD':
        return row['quantity'] * 1
    elif row['action'] == 'DELETE':
        return row['quantity'] * -1
    else:
        return row['quantity'] * 1

df['moved_quantity_to'] = df.apply(stock, axis=1)

df
```

```python
df_from = df[['id','from_location_id','moved_quantity_from','material_name']]
df_from
```

```python
df_from = df_from.rename(columns={"from_location_id":"location","moved_quantity_from":"quantity_moved"})
df_from
```

```python
df_to = df[['id','to_location_id','moved_quantity_to','material_name']]
df_to
```

```python
df_to = df_to.rename(columns={"to_location_id":"location","moved_quantity_to":"quantity_moved"})
df_to = df_to[df_to['location'].isnull() == False]
df_to
```

```python
all_df = pd.concat([df_from,df_to])
all_df
```

```python
time = pd.read_csv('timestamps.csv')
time['timestamp'] = pd.to_datetime(time['timestamp'])
time
```

```python
merged = all_df.merge(time,on='id',how='inner')
merged.count()
```

```python
def movement(row):
    if row['quantity_moved'] >= 0:
        return  'in'
    elif row['quantity_moved'] < 0: 
        return 'out'

merged['movement'] = merged.apply(movement,axis=1)
merged

```

```python
merged.to_csv('raw.csv')
```

```python
r = pd.pivot_table(merged, values='quantity_moved',columns='movement',index=['material_name','timestamp'],aggfunc="sum")
r
```

```python
r = pd.pivot_table(merged, values='quantity_moved',columns='movement',index=['material_name','timestamp'],aggfunc="sum")
r.to_csv('material_movements.csv') # para ajustar los nombres de las columnas
r = pd.read_csv('material_movements.csv')
r
```
