```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('material_movements.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['month'] = df['timestamp'].dt.month
df['year'] = df['timestamp'].dt.year
```

```python
df['date'] = df['month'].astype('str') + '-' + df['year'].astype('str')
df = df.drop(['month','year','timestamp'],axis=1)
df
```

```python
df1 = df.groupby(by=['date','material_name']).sum().sort_values(['in','date'],ascending=True)
df1
```

```python
df1.to_csv('temp.csv')
df1 = pd.read_csv('temp.csv')
df1['total'] = df1['in'] + df1['out']
df1
```

```python
def move(row):
    if row['total'] == 0:
        return 'good'
    elif row['total'] < 1:
        return 'below'
    elif row['total'] > 1:
        return 'above'

df1['move'] = df1.apply(move,axis=1)
df1
```

```python
df2 = pd.read_csv('material_frequency.csv')
df3 = df1.merge(df2,on='material_name',how='inner',)
df3 = df3.drop(['hola'],axis=1)
df3.sort_values(['date'])
```
