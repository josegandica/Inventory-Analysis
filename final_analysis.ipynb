```python
import pandas as pd
import numpy as np
```

```python
## Import timestamps
time = pd.read_csv('timestamps.csv')
##transform timestamps to datetime
time['timestamp'] = pd.to_datetime(time['timestamp'])
## Import activities table
df = pd.read_csv('main.csv')
##correct quantity data type
df['quantity'] = df['quantity'].str.replace(',','')
df['quantity'] = pd.to_numeric(df['quantity'])
##rename columns of locations
df = df.rename(columns={'from_location_id':'from','to_location_id':'to'})
## merge df with time
df = df.merge(time,on='id',how='inner')
df 
```

```python
r = df

def movement_from(row):
    if row['action'] =='ADD':
        return 'IN'
    elif row['action'] == 'TRANSFER':
        return 'OUT'
    elif row['action'] == 'DELETE':
        return 'OUT'
r['from_movement'] = r.apply(movement_from,axis=1)

```

```python
r = r.loc[r['from_movement'] == 'OUT']
r['month'] = r['timestamp'].dt.month
r['year'] = r['timestamp'].dt.year
```

```python
r['date'] = r['month'].astype('str') + '-' + r['year'].astype('str')
r
```

```python
r = r.drop(columns=['id','to','action','timestamp','from_movement','month','year'])
```

```python
r['date'] = pd.to_datetime(r['date'])
```

```python
r = r.groupby(by=['from','material_name','date']).sum().reset_index()
r
```

```python
min = r.groupby(by=['from','material_name']).min().reset_index()
min.drop(columns='date',axis=1)
```

```python
max = r.groupby(by=['from','material_name']).mean().reset_index()
max['max'] = np.ceil(max['quantity'] * 1.3)
max
```

```python
min = min.rename(columns={'quantity':'min'})

```

```python
final = min.merge(max,on=['from','material_name'])
final = final.drop(columns='date',axis=1)
final
```

```python
final.to_csv('final_Kyle Plumbing.csv')
```
